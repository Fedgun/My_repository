```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

driver_path = r'D:\IDE_SF\chromedriver.exe'


def close_annoying_banners(driver):
    
    # Закрытие баннеров
    xpaths_to_try = [
        "//button[contains(., 'Хорошо')]",
        "//button[contains(., 'Понятно')]",
        "//button[contains(., 'Соглас')]",
        "//button[contains(., 'Принять')]",

        "//button[contains(., 'Не сейчас')]",
        "//button[contains(., 'Закрыть')]",
        "//div[contains(@class,'tp-quiz')]//button[contains(., 'Закрыть')]",

        "//button[contains(., 'Отказаться')]",
        "//button[contains(., 'Разрешить')]",
    ]

    for xp in xpaths_to_try:
        try:
            btn = driver.find_element(By.XPATH, xp)
            driver.execute_script(
                "arguments[0].scrollIntoView({block: 'center'});",
                btn
            )
            btn.click()
        except Exception:
            pass


def run_quiz(driver_path):
    service = Service(driver_path)
    driver = webdriver.Chrome(service=service)
    wait = WebDriverWait(driver, 10)

    driver.get('https://tproger.ru/quiz/real-programmer')

    while True:
        close_annoying_banners(driver)

        # ждём сам вопрос
        question_el = wait.until(
            EC.presence_of_element_located(
                (By.CSS_SELECTOR, ".tp-quiz__question.tp-quiz-question")
            )
        )
        question_text = question_el.text.strip()

        # варианты
        answer_items = wait.until(
            EC.presence_of_all_elements_located(
                (By.CSS_SELECTOR, ".tp-quiz-answer-list__item")
            )
        )

        # Повторное закрытие мусора на всякий случай
        close_annoying_banners(driver)

        # берём первый ответ (label с радио)
        first_answer_label = answer_items[0].find_element(By.CSS_SELECTOR, ".tp-answer__label")

        # скроллим его в зону клика
        driver.execute_script(
            "arguments[0].scrollIntoView({block: 'center'});",
            first_answer_label
        )

        close_annoying_banners(driver)

        first_answer_label.click()

        # ждём правильный ответ
        correct_block = wait.until(
            EC.presence_of_element_located(
                (By.CSS_SELECTOR, ".tp-answer--result-true")
            )
        )
        correct_text = correct_block.find_element(
            By.CSS_SELECTOR,
            ".tp-answer__text"
        ).text.strip()

        print("Вопрос:", question_text)
        print("Правильный ответ:", correct_text)
        print("-" * 40)


        # ищем кнопку "Далее"
        try:
            next_btn = wait.until(
                EC.element_to_be_clickable(
                    (By.XPATH, "//button[contains(., 'Далее')]")
                )
            )
            driver.execute_script(
                "arguments[0].scrollIntoView({block: 'center'});",
                next_btn
            )
            close_annoying_banners(driver)
            next_btn.click()
        except Exception:
            # если не нашли/не кликнули - значит вопросов больше нет
            break

    driver.quit()


run_quiz(driver_path)
```
